```python
#匯入在sklearn.linear_model套件裡面的LinearRegression模型
#匯入在sklearn.model_selection套件裡面的train_test_split模組
#切割數據集(訓練資料80%、測試資料20%，設定random_state=20)
#畫出訓練資料集的matplotlib圖形
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split


# 1. x 為 2D 格式
x_2d = np.array(x).reshape(-1, 1)

# 2. 切割數據集(訓練資料80%、測試資料20%，設定random_state=20)
x_train, x_test, y_train, y_test = train_test_split(x_2d, y, test_size=0.2, random_state=20)

# 3. 畫出訓練資料集的matplotlib圖形
plt.scatter(x_train, y_train, color='blue', label='Training Data')
plt.title('Training Data Visualization')
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.legend()
plt.show()
```

## 用訓練資料來 fit 函數
1. 只用訓練資料集的資料進行linear regression演算法<br>
2. 並計算出訓練階段的MSE
3. 畫出目標點(藍色)與預測點(紅色)的對應圖
![](images/LR_train.PNG)

```python
from sklearn.metrics import mean_squared_error


# 1. 只用訓練資料集的資料進行linear regression演算法
model = LinearRegression()
model.fit(x_train, y_train)
print('Coefficients: ', model.coef_)
print('interception: ', model.intercept_)

# 2. 預測訓練集自己的結果
y_train_pred = model.predict(x_train)

# 3. 計算訓練階段的 MSE (均方誤差)
mse_train = mean_squared_error(y_train, y_train_pred)
print(f"訓練集的 MSE: {mse_train}")
```

```python
# 1. 取得預測結果 (使用訓練資料的 x 來產生對應的預測 y)
y_train_pred = model.predict(x_train)

# 2. 畫出原始目標點 (藍色散佈圖)
plt.scatter(x_train, y_train, color='blue', label='Actual (Target)')

# 3. 畫出預測點/線 (紅色直線)
plt.plot(x_train, y_train_pred, color='red', linewidth=2, label='Predicted (Regression Line)')

# 圖表設定
plt.title('Actual vs Predicted (Training Set)')
plt.xlabel('X (Feature)')
plt.ylabel('Y (Target)')
plt.legend()
plt.show()
```

## 將訓練出來的函數預測測試集的X值
1. 使用剛剛訓練出來的模型進行測試資料集的資料預測*注意reshape<br>

```python
#1. 單點預測測試：X = 2.44897959

target_x = np.array([[2.44897959]])
single_pred = model.predict(target_x)

print(f"單點 X=2.44897959 的預測結果: {single_pred[0]}")
```

```python
# 2. 對所有測試集資料進行預測
# 這裡的 x_test 已經在前面切割時 reshape 過了

y_test_pred = model.predict(x_test)

print("測試集前五筆預測結果：", y_test_pred[:5])
```

2. 並計算出測試階段的MSE

```python
from sklearn.metrics import mean_squared_error

# 1. 使用測試集 X (x_test) 進行預測
y_test_pred = model.predict(x_test)

# 2. 計算測試階段的 MSE (比較 真實答案 y_test 與 預測答案 y_test_pred)
mse_test = mean_squared_error(y_test, y_test_pred)

print(f"測試集的 MSE: {mse_test}")
```
