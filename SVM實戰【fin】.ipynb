# 乳癌資料庫預測SVM分類
>使用scikit-learn 機器學習套件裡的SVR演算法

* (一)引入函式庫及內建乳癌資料集<br>
引入之函式庫如下<br>
sklearn.datasets: 用來匯入內建之乳癌資料集`datasets.load_breast_cancer()`<br>
sklearn.SVR: 支持向量機回歸分析之演算法<br>
matplotlib.pyplot: 用來繪製影像

````python
from sklearn import svm
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
````

## Step1. 下載資料

````python
breast_cancer=datasets.load_breast_cancer()
````

````python
features=breast_cancer.data
features
````

````python
target=breast_cancer.target
target
````

````python
print(breast_cancer.keys())
print(breast_cancer.feature_names)
print(breast_cancer.target_names)
````

````python
target_names=breast_cancer.target_names
labels=target_names[target]
print(target_names)
print(labels)
````

## Step2. 區分訓練集與測試集

### 📊 SVM 與資料切分核心參數一覽表


| 參數名稱 | 所屬步驟 | 技術定義 | 🎯 一目了然的白話意義 | 調校影響 / 實務建議 |
| :--- | :--- | :--- | :--- | :--- |
| **`test_size`** | 資料切分 | 測試集佔總資料的比例 | **保留多少比例的資料來考試** | 常用 `0.2` 或 `0.3`。過大模型學不飽，過小無法準確評估。 |
| **`random_state`** | 資料切分 | 隨機數生成器的種子碼 | **鎖定洗牌順序（錄影重播鍵）** | 數字大小無所謂。固定它才能**讓每次實驗的對比基準都一樣**。 |
| **`stratify`** | 資料切分 | 分層抽樣依據（通常帶入標籤 `y`）| **確保切分前後各類別比例不變** | 極重要！防止隨機抽樣時，某個類別被漏抽或集中在某一邊。 |
| **`kernel`** | 模型建立 | 決定轉換資料空間的核函數 | **切分資料的數學手法（直線或曲線）** | `"linear"` 適用簡單線性分類；`"rbf"`（預設）適用錯綜複雜的非線性邊界。 |
| **`C`** | 模型建立 | 錯誤樣本的懲罰係數 | **模型對分類錯誤的嚴格程度** | **越大越嚴格**：不容許出錯，邊界變窄，容易過擬合。<br>**越小越寬容**：允許微幅出錯，邊界變寬，泛化力較佳。 |
| **`gamma`** | 模型建立 | RBF 核函數的影響半徑 (預設 `scale`) | **單一資料點影響邊界的磁場範圍** | **越大越短視**：只看附近的點，邊界扭曲變形（易過擬合）。<br>**越小越遠矚**：看整體的點，邊界平滑大器（易欠擬合）。 |

````python
x_train, x_test, y_train, y_test = train_test_split(
    features, labels, test_size=0.3, random_state=42, stratify=labels
)
````

## Step3. 建模

````python
#資料縮放
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
````

````python
#建模
clf = svm.SVC(kernel="linear", C=1.0)  # 使用線性核函數
clf.fit(x_train_scaled, y_train)
````

## Step4. 預測

```

```


````python
y_pred = clf.predict(x_test_scaled)
````

````python
y_pred
````

## Step5. 準確度分析

````python
accuracy = accuracy_score(y_test, y_pred)
````

````python
print(f"訓練集樣本數: {x_train.shape[0]}")
print(f"測試集樣本數: {x_test.shape[0]}")
print(f"測試集預測準確度: {accuracy:.2%}")
````
